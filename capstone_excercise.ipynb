```python
import seaborn as sns
fig, ax = plt.subplots(figsize=(8,5))

df_100=df.head(100)
sns.scatterplot(data=df_100, x="distance", y="fare")

ax.set_title('Trip Distance vs Fare Amount')

ax.set_xlabel('distance (KM)')

ax.set_ylabel('fare')

plt.show()



```

```python
fig, ax = plt.subplots(figsize=(6, 4))
df_100=df.head(100)
# Bar charts look clean and well-spaced when X is a true category with only a few values
sns.barplot(data=df_100, x='payment', y='total')

ax.set_title('Average Total Amount Paid by Payment Type')
ax.set_xlabel('Payment Method')
ax.set_ylabel('Average Total Amount ($)')

plt.tight_layout()
plt.show()
```

```python
# Requirement fulfilled: plt.subplots with 2 panels side-by-side
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
df_100=df.head(100)
# Histogram of Distances (Type 3 Chart)
sns.histplot(data=df_100, x='distance', bins=10, color='#2ca02c', kde=True, ax=axes[0])
axes[0].set_title('Distribution of Trip Distances')
axes[0].set_xlabel('Trip Distance (Miles)')
axes[0].set_ylabel('Frequency')

#  Box Plot of Passenger Counts (Type 4 Chart)
sns.boxplot(data=df_100, x='passengers', color='#ff7f0e', ax=axes[1])
axes[1].set_title('Spread of Passenger Counts per Trip')
axes[1].set_xlabel('Number of Passengers')

plt.tight_layout()
plt.show()
```
